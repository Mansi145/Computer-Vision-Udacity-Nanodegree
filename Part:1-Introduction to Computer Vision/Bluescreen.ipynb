```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

%matplotlib inline
```

```python
path = "..\Part 01-Module 01-Lesson 02_Image Representation  Classification\img"
```

```python
os.chdir(path)
```

```python
image = cv2.imread("pizza_bluescreen.jpg")

img_copy = np.copy(image)
img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)

plt.imshow(img_copy)
```

```python
lower_blue = np.array([0,0,200])
upper_blue = np.array([200,200,255])
```

```python
mask = cv2.inRange(img_copy, lower_blue, upper_blue)
plt.imshow(mask, cmap = "gray")
```

```python
original_img = np.copy(img_copy)

original_img[mask !=0 ] = [0,0,0]

plt.imshow(original_img)
```

```python
background = cv2.imread("sky.jpg")

background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)

cropped_background = background[:514, :816]

cropped_background[mask == 0] = [0,0,0]

plt.imshow(cropped_background)
```

```python
complete_img = cropped_background + original_img

plt.imshow(complete_img)
```
